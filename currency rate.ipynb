```python
# Example of converting string date to timestamp
from datetime import datetime

# Method 1: Using strptime() to parse string date
date_str = "2025-01-02"
dt = datetime.strptime(date_str, "%Y-%m-%d")
timestamp = int(dt.timestamp())

print(f"String date: {date_str}")
print(f"Datetime: {dt}")
print(f"Timestamp: {timestamp}")

# Currency rate for the date
rate_eur = cnb_exchange_rate.daily_rate('EUR', datetime.fromtimestamp(timestamp))
print(f"Currency rate: {rate_eur}")

# Method 2: If you have day, month, year as separate strings
year, month, day = "2025", "01", "02"
dt2 = datetime(int(year), int(month), int(day))
timestamp2 = int(dt2.timestamp())

print(f"\nComponents: year={year}, month={month}, day={day}")
print(f"Datetime: {dt2}")
print(f"Timestamp: {timestamp2}")

# Convert timestamp back to datetime to verify
print(f"\nTimestamp back to datetime: {datetime.fromtimestamp(timestamp)}")
```
